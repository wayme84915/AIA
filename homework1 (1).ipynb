```python
train = train[train['type']!=4]
#除去 type=4的值
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = train[['花瓣寬度','花瓣長度','花萼寬度','花萼長度']]
y = train['type']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state = 100, shuffle=True)
#訓練的值
```

```python
X_train
```

```python
from sklearn.svm import SVC
from sklearn import metrics
```

```python
model = SVC()
model.fit(X_train, y_train)
prediction = model.predict(X_test).astype(np.int)
print("accuracy:", model.score(X_train, y_train))
#使用SVC為模型,並進行訓練
```

```python
test1 = test[['花瓣寬度','花瓣長度','花萼寬度','花萼長度']]
svm_predicted = model.predict(test1).astype(np.int)
```

```python
svm_predicted 
#觀測結果
```

```python
submission = pd.read_csv('submission.csv', encoding = "utf-8", dtype = {'type': np.int32})
```

```python
StackingSubmission = pd.DataFrame({'id': submission.id, 'type':svm_predicted  })
StackingSubmission.to_csv("submission.csv", index = False)
```
